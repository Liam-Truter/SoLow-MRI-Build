# Illustrating Bore Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from mpl_toolkits.mplot3d import Axes3D

# Bore parameters
radius = 10  # mm (change to 164)
height = 10  # mm (change to 436) 
spacing = 1   # mm, distance between points

# Generate points in cylindrical coordinates
x_vals = np.arange(-height / 2, height / 2 + spacing, spacing)  # along x-axis (height direction)
theta_vals = np.arange(0, 2 * np.pi, spacing / radius)  # angular direction
r_vals = np.arange(0, radius + spacing, spacing)  # radial direction

# Hold Cartesian coordinates
X = []
Y = []
Z = []

# Cylindrical coordinates to Cartesian
for x in x_vals:
    for r in r_vals:
        for theta in theta_vals:
            X.append(x)
            Y.append(r * np.cos(theta))
            Z.append(r * np.sin(theta))

# Numpy arrays for easier plotting
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)

# Save files to csv
with open('cylinder_points.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['x', 'y', 'z'])  # CSV header
    for i in range(len(X)):
        writer.writerow([X[i], Y[i], Z[i]])

print("Coordinates saved to 'cylinder_points.csv'")

# Plotting the points in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, s=1, color='red', alpha=0.6)

# Plot limits and labels
ax.set_xlabel('X (Height)')
ax.set_ylabel('Y (Radius)')
ax.set_zlabel('Z (Radius)')
ax.set_title("Points inside a Cylinder (20 mm diameter, 10 mm height)")

# Show the plot
plt.show()
```
